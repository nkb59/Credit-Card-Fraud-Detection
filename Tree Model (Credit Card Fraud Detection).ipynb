```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, f1_score, roc_auc_score, precision_score, recall_score
```

```python
credit_card_data = pd.read_csv("/Users/niketanbaranwal/Downloads/creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.isnull().sum()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
```

```python
X = credit_card_data.drop('Class',axis=1)
y = credit_card_data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
dt = DecisionTreeClassifier(criterion='entropy', random_state = 42)
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
fig = plt.figure(figsize=(25,20))
_ = plot_tree(dt,
                   filled=True)
```

```python
print("Decision Tree without Class Weights:")
print(classification_report(y_test, y_pred))
```

```python
print(f"Accuracy Score: {accuracy_score(y_test,y_pred)}")
print(f"Precision Score: {precision_score(y_test,y_pred)}")
print(f"Recall Score: {recall_score(y_test,y_pred)}")
print(f"F1-Score: {f1_score(y_test, y_pred)}")
print(f"AUC: {roc_auc_score(y_test, dt.predict_proba(X_test)[:, 1])}")
```

```python
matrix_tree = confusion_matrix(y_test, y_pred)
cm_tree = pd.DataFrame(matrix_tree, index=['not_fraud', 'fraud'], columns=['not_fraud', 'fraud'])

sns.heatmap(cm_tree, annot=True, cbar=None, cmap="Greens", fmt = 'g')
plt.title("Confusion Matrix for Decision Tree without Weights"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

```python
#Using balanced class weights
dt_weighted = DecisionTreeClassifier(class_weight='balanced',criterion='entropy', random_state=42, max_depth = 4)
dt_weighted.fit(X_train, y_train)
y_pred_weighted = dt_weighted.predict(X_test)
```

```python
fig = plt.figure(figsize=(20,20))
_ = plot_tree(dt_weighted,
                   filled=True)
```

```python
print("Decision Tree with Class Weights:")
print(classification_report(y_test, y_pred_weighted))
```

```python
print(f"Accuracy Score: {accuracy_score(y_test,y_pred_weighted)}")
print(f"Precision Score: {precision_score(y_test,y_pred_weighted)}")
print(f"Recall Score: {recall_score(y_test,y_pred_weighted)}")
print(f"F1-Score: {f1_score(y_test, y_pred_weighted)}")
print(f"AUC: {roc_auc_score(y_test, dt_weighted.predict_proba(X_test)[:, 1])}")
```

```python
matrix_tree = confusion_matrix(y_test, y_pred_weighted)
cm_tree = pd.DataFrame(matrix_tree, index=['not_fraud', 'fraud'], columns=['not_fraud', 'fraud'])

sns.heatmap(cm_tree, annot=True, cbar=None, cmap="Purples", fmt = 'g')
plt.title("Confusion Matrix for Decision Tree with Weights"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate a Random Forest Classifier
# You can adjust the hyperparameters based on your needs and data

random_forest_model = RandomForestClassifier(
    n_estimators=100,      # Number of trees in the forest (common starting point)
    criterion='entropy',       # Splitting criterion ('gini' or 'entropy')
    max_depth=None,         # Maximum depth of the trees (None means no limit, but can lead to overfitting)
    min_samples_split=2,    # Minimum number of samples required to split an internal node
    min_samples_leaf=1,     # Minimum number of samples required to be at a leaf node
    max_features='sqrt',    # Number of features to consider when looking for the best split ('sqrt', 'log2', None, or int/float)
    bootstrap=True,         # Whether bootstrap samples are used when building trees
    oob_score=True,         # Whether to estimate out-of-bag score
    random_state=42,        # For reproducibility
    class_weight='balanced', # To handle class imbalance (important for fraud detection)
    n_jobs=-1              # Use all available CPU cores for training
)
```

```python
random_forest_model.fit(X_train, y_train)

y_pred_rf = random_forest_model.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
roc_auc_rf = roc_auc_score(y_test, random_forest_model.predict_proba(X_test)[:, 1])
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
class_report_rf = classification_report(y_test, y_pred_rf)

print("Random Forest Metrics:")
print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1-Score: {f1_rf:.4f}")
print(f"AUC: {roc_auc_rf:.4f}")
print("Confusion Matrix:\n", conf_matrix_rf)
print("Classification Report:\n", class_report_rf)
```
